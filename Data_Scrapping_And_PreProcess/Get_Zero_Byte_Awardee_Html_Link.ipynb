```python
missing_ids_by_file = []
target_file = open("html_links_for_zero_byte_awardees.txt", 'w')

with open("ZERO_BYTE_ID.txt", "rb") as f:
    missing_ids = []
    urlfile_index_previous = 1
    for i, line in enumerate(f):
        urlfile_index = int(line) / 3500
        
        if int(line) % 3500 != 0:
            urlfile_index += 1
            
        if urlfile_index == urlfile_index_previous:
            missing_ids.append(line)
        else:
            missing_ids.append("-1")
            missing_ids_by_file.append(missing_ids)
            missing_ids = []
            missing_ids.append(line)
            urlfile_index_previous += 1            
    missing_ids.append("-1")
    missing_ids_by_file.append(missing_ids)        
    f.close()    

# sum = 0
# for i in xrange(0, len(missing_ids_by_file)):
#     print str(i)+" "+ str(len(missing_ids_by_file[i]))
#     sum += len(missing_ids_by_file[i-1])
# print sum
    
numberOfWrittenLines = 1
for i in xrange(0, 21):
    print str(i) + " " + str(len(missing_ids_by_file[i]))
    urlfile = "all-authors-html/splited-21-files/authors_urls" + str(i+1) + ".txt"
    with open(urlfile, "rb") as urls:
        print "opening - " + urlfile
        missing_line_file_counter = 0
        print "i = " + str(i)
        print "len(missing_ids_by_file[i]) = " + str(len(missing_ids_by_file[i]))
        for index, line in enumerate(urls):
            missing_line = missing_ids_by_file[i][missing_line_file_counter] 
            if index == int(missing_line)%3500:
                numberOfWrittenLines += 1
                target_file.write(line.encode('utf-8'))
#                 target_file.write("\n")
                missing_line_file_counter += 1
        missing_line_file_counter = 0
        
print "Number Of Written URLs for Missing Awardees = " + str(numberOfWrittenLines)
target_file.close()
```
